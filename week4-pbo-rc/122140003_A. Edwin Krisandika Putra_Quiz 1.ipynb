```python
import random
listkata = [
    'algorithm', 'binary', 'boolean', 'byte', 'cache', 'compiler', 'debugger',
    'encryption', 'framework', 'function', 'garbage', 'hash', 'index', 'iterator',
    'javascript', 'json', 'library', 'loop', 'namespace', 'object', 'operator',
    'overload', 'polymorphism', 'queue', 'recursion', 'serialization', 'stack',
    'template', 'variable', 'virtual', 'web', 'xml', 'yaml', 'zip'
]

def pilihKata():
    return random.choice(listkata)
```

```python
def hangman(sisaPercobaan):
    visual = [
        """
            ------
            |    |
            |    O
            |  --|--
            |    |
            |   / \\
            |
        ------------
        """,
        """
            ------
            |    |
            |    O
            |  --|
            |    |
            |   / \\
            |
        ------------
        """,
        """
            ------
            |    |
            |    O
            |    |
            |    |
            |   / \\
            |
        ------------
        """,
        """
            ------
            |    |
            |    O
            |    |
            |    |
            |   /
            |
        ------------
        """,
        """
            ------
            |    |
            |    O
            |    |
            |    |
            |
            |
        ------------
        """,  
        """
            ------
            |    |
            |    O
            |
            |
            |
            |
        ------------
        """,
        """
            ------
            |    |
            |
            |
            |
            |
            |
        ------------
        """
        ]
    
    return visual[sisaPercobaan]
```

```python
def program(kata):
    hurufDiTebak = []
    tebakanKata = []
    progress = len(kata) * "_"
    sisaPercobaan = 6
    
    tebakanIsTrue = False
    print(hangman(sisaPercobaan))
    while sisaPercobaan > 0 and not tebakanIsTrue:
        print(progress)
        print("\n")
        tebak = input("Masukkan sebuah huruf : ").lower()
        if len(tebak) == 1 and tebak.isalpha():
            if tebak in hurufDiTebak:
                print(f"Huruf {tebak} sudah pernah dipilih!")
            elif tebak not in kata:
                sisaPercobaan -= 1
                print(f"{tebak} salah! Sisa Percobaan : {sisaPercobaan} ")
                hurufDiTebak.append(tebak)
                print(hangman(sisaPercobaan))

            else:
                print(f"Benar! {tebak} termasuk dalam kata.")
                hurufDiTebak.append(tebak)

                cetakKata = list(progress)
                index = []
                for i, huruf in enumerate(kata):
                    if huruf == tebak:
                        index.append(i)

                for i in index:
                    cetakKata[i] = tebak
                progress = "".join(cetakKata)
                
                if "_"not in progress:
                    tebakanIsTrue = True
        else:
            print(hangman(sisaPercobaan))
            print(progress)
            print("Masukkan tebakan tidak valid!\n")
    
    if tebakanIsTrue:
        print(f"Selamat! Anda berhasil menebak kata : {kata}")
    else:
        print(f"Percobaan habis! Orangnya matik, kata yang benar adalah {kata}.")
```

```python
def play():
    while True:
        kata = pilihKata()
        program(kata)
        
        if input("Main lagi? (y/n)").lower() != "y":
            break
```

```python
play()
```
